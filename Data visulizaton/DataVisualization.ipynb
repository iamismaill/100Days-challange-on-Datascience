Line Chart

```python
## Line Chart 
  ## Line plot is a chart that displays a line between two variables 
import matplotlib.pyplot as plt 

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3] 
plt.plot(years,gdp,color='blue',marker='*',linestyle='solid')
plt.title("GDP per Year")
plt.xlabel("Years")
plt.style.use("seaborn-colorblind")
plt.ylabel("Billions of $")
plt.show()
```

Bar Charts 
* bar chart is good choice when we want to show quantity varies among some discrte set of items for exampe how many academy awards were won by of variety of movies 

```python
##Bar chart 

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

#in order to use bar char we have to use the function of plt.bar()

plt.bar(movies,num_oscars)
plt.title("My Favourite Movies")
plt.ylabel("Academy of awards")
plt.xlabel("Movies")
plt.show()
```

```python
### Bar plot 
movie_ratings =[1,2,3,4,5]
rating_counts =[20,40,50,70,60]
plt.bar(movie_ratings,rating_counts,color='#bc5090')
plt.title("Movie Rating Distribution")
plt.xlabel("Movie Rating")
plt.ylabel("Rating Frequency")
plt.show()
```

```python
## Bar chart showing tech companies revenue per year
tech_companies = ['Google','Micrsoft','Facebook','Amazon','Instagram','Uber']
revenue = [90,80,78,70,123,200]
plt.bar(tech_companies,revenue)
plt.title("Tech Companies Revenue")
plt.xlabel("Tech Companies")
plt.ylabel("Revenue (blns) ")
plt.show()
```

```python
## Bar chart showing tech companies revenue per year

tech_companies = ['Google','Micrsoft','Facebook','Amazon','Instagram','Uber']
profit = [50,40,38,45,100,145]
revenue = [90,80,78,70,123,200]
plt.bar(tech_companies,revenue,label='Revenue')
plt.bar(tech_companies,profit,color ='green',label='Profit')
plt.title("Tech Companies Revenue")
plt.xlabel("Tech Companies")
plt.ylabel("Revenue (blns) ")
##generates the labels 
plt.legend()
plt.show()
```

```python
## Side by side revenu and profit for tech companies

##if we want to make side by side we have to convert our xlabels into numpy format 

import numpy as np 
tech_companies = ['Google','Micrsoft','Facebook','Amazon','Instagram','Uber']
xpos = np.arange(len(tech_companies))
profit = [50,40,38,45,100,145]
revenue = [90,80,78,70,123,200]
plt.bar(xpos +0.2,revenue,width= 0.3, label='Revenue')
plt.xticks(xpos,tech_companies)
plt.bar(xpos -0.2 ,profit,width=0.3,color ='green',label='Profit')
plt.title("Tech Companies Revenue")
plt.xlabel("Tech Companies")
plt.ylabel("Revenue (blns) ")
##generates the labels 
plt.legend()
plt.show()
```

Pie Plot : is circular graph that splitup into wedge shaped , each piece is propportionate to the value it represents 

 PIE CHARTS

```python
##Pie Chart 

## Sample Data 
subjects = ['C++','Java','Python','R','Data Mining']
marks = [87,80,70,98,56]
plt.pie(marks,
labels=subjects,
colors=['#ff6361','#bc5090','y','g','#003f5c'],
startangle=20,
shadow=True,
explode=(0,0.1,0,0,0),
autopct='%1.1f%%')
plt.title("Student Performance")
```

```python
## Pie Chart in Dataframe 
drinks = ['Milk Tea','Green Tea','Burble Tea','Black Tea','Arabic Tea']
data = [70,40,65,90,65]
plt.pie(data , 
labels =drinks,
colors=['#2f4b7c','#665191','#d45087','#ff7c43','#9dc6e0'],
shadow=True,
explode=[1,0,0,0,0],

autopct='%1.1f%%')


 
# Adding legend
plt.legend(drinks,
          title ="Drinks",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
plt.title("Comparing Tea Drinks")
```

Histogram Plot : Shows the distribution of numeric variable , we create histogram using hist() method 

```python
##Histogram plot 
employee_age =[21,26,29,34,36,39,40,43]
##bins
bins = [20,30,40,50,60]
##Ploting the histogram
plt.hist(employee_age,bins,rwidth=0.6)
plt.xlabel("Employee Age")
plt.ylabel("Frequency")
plt.title("Employee Age Distribution")
plt.show()
```

LM Plots are like scatter plots , Scatter plots is the best way to understand the relationship between two variables such salary vs age
LM plots take two column names - x and y as string 

```python
import pandas as pd 
data = pd.DataFrame({'x':[1,3,5,7,9,11],'y':[10,25,35,33,41,59]})
import seaborn as sns
sns.lmplot(x='x',y='y',data=data)
```
